```python
import pandas as pd
path = '/content/melb_data.csv'
home_data = pd.read_csv('train.csv') # or path
#df.describe()
#home_data.columns
```

```python
#home_data = pd.read_csv('melb_data.csv')
home_data
```

```python
from sklearn.tree import DecisionTreeRegressor as DTR
home_model = DTR(random_state=1)
feature_names = ['LotArea','1stFlrSF','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = home_data[feature_names]
y = home_data['SalePrice']
home_model.fit(X, y)

```

MAE : mean absolute error

error = active value - prediction velue

```python
predictions = home_model.predict(X)
print(predictions)
```

```python
from sklearn.metrics import mean_absolute_error as mae
mae(y,predictions)
```

```python
from sklearn.model_selection import train_test_split as tts
train_X, val_X, train_y, val_y = tts(X,y,random_state=0)
home_model2=DTR()
home_model2.fit(train_X,train_y)
val_perdt=home_model2.predict(val_X)
mae(val_y,val_perdt)
```
